```python
import torch
import numpy as np
from transformers import VitsModel, AutoTokenizer
import soundfile as sf

# 1. DETECTOR DE HARDWARE
device = "cuda" if torch.cuda.is_available() else "cpu"
dtype = torch.float32
print(f"🚀 Entorno listo. Usando: {device}")

# 2. CARGA DEL MODELO TTS (ESPAÑOL)
model_id = "facebook/mms-tts-spa"

print("Cargando modelo TTS en memoria... (esto puede tardar)")
tokenizer = AutoTokenizer.from_pretrained(model_id)
model = VitsModel.from_pretrained(model_id).to(device)

# 3. TEXTO A CONVERTIR
text = "Hola, este es un ejemplo de síntesis de voz usando inteligencia artificial en español."

# 4. TOKENIZACIÓN
inputs = tokenizer(text, return_tensors="pt").to(device)

# 5. GENERACIÓN DE AUDIO
print("Generando audio...")
with torch.no_grad():
    output = model(**inputs).waveform

# 6. PROCESAMIENTO Y GUARDADO
audio = output.cpu().numpy().squeeze()

sf.write("resultado_tts.wav", audio, samplerate=16000)

print("Listo, audio guardado como resultado_tts.wav")
```
